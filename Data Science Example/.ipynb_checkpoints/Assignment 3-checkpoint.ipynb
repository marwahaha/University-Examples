```python
"""
tree69.py
A tree utility library for COMPSCI 369.
"""

# The MIT License (MIT)
#
# Copyright (c) 2016 Arman Bilge and Stuart Bradley
#
# Permission is hereby granted, free of charge, to any person obtaining a copy
# of this software and associated documentation files (the "Software"), to deal
# in the Software without restriction, including without limitation the rights
# to use, copy, modify, merge, publish, distribute, sublicense, and/or sell
# copies of the Software, and to permit persons to whom the Software is
# furnished to do so, subject to the following conditions:
#
# The above copyright notice and this permission notice shall be included in all
# copies or substantial portions of the Software.
#
# THE SOFTWARE IS PROVIDED "AS IS", WITHOUT WARRANTY OF ANY KIND, EXPRESS OR
# IMPLIED, INCLUDING BUT NOT LIMITED TO THE WARRANTIES OF MERCHANTABILITY,
# FITNESS FOR A PARTICULAR PURPOSE AND NONINFRINGEMENT. IN NO EVENT SHALL THE
# AUTHORS OR COPYRIGHT HOLDERS BE LIABLE FOR ANY CLAIM, DAMAGES OR OTHER
# LIABILITY, WHETHER IN AN ACTION OF CONTRACT, TORT OR OTHERWISE, ARISING FROM,
# OUT OF OR IN CONNECTION WITH THE SOFTWARE OR THE USE OR OTHER DEALINGS IN THE
# SOFTWARE.

def compute_upgma_tree(matrix):

    import itertools as it

    n = len(matrix)
    nodes = [Node(str(i + 1)) for i in range(n)]
    for node in nodes:
        node.set_height(0)
    matrix = {nodes[i]: {nodes[j]: matrix[i][j] for j in range(n)} for i in range(n)}
    while len(matrix) > 1:
        a, b = min(it.combinations(matrix.keys(), 2), key=lambda xy: matrix[xy[0]][xy[1]])
        u = Node()
        u.add_child(a)
        u.add_child(b)
        u.set_height(matrix[a][b] / 2)
        uc = {c: (a.get_leaf_count() * matrix[a][c] + b.get_leaf_count() * matrix[b][c]) / (a.get_leaf_count() + b.get_leaf_count()) for c in matrix.keys() - set((a, b))}
        del matrix[a]
        del matrix[b]
        for k, v in matrix.items():
            del v[a]
            del v[b]
            v[u] = uc[k]
        matrix[u] = uc
    return Tree(u)

def plot_tree(tree):

    import itertools as it
    import numpy as np
    from matplotlib import pyplot as plt

    def compute_node_xy(node, counter=it.count()):
        node.x = node.get_height()
        if node.is_leaf():
            node.y = next(counter)
        else:
            children = node.get_children()
            for child in children:
                compute_node_xy(child, counter)
            node.y = np.mean([c.y for c in children])

    def plot_node(node):
        if node.is_leaf():
            plt.text(node.x, node.y, ' ' + node.get_label(), {'ha':'left', 'va':'center'})
        else:
            children = node.get_children()
            plt.plot([node.x] * 2, [min(c.y for c in children), max(c.y for c in children)], 'k')
            for child in children:
                plt.plot([node.x, child.x], [child.y] * 2, 'k')
                plot_node(child)

    root = tree.get_root()
    compute_node_xy(root)
    plt.plot([root.x, root.x + root.x/16], [root.y] * 2, 'k')
    plot_node(root)
    lc = tree.get_leaf_count()
    plt.ylim(- lc / 16, 17/16 * lc - 1)
    axes = plt.gca()
    axes.invert_xaxis()
    axes.yaxis.set_visible(False)
    axes.set_frame_on(False)
    axes.grid()
    plt.xlabel("Ne")
    plt.ylabel("Leaves")
    plt.show()

"""
Python Tree Class
Stuart Bradley - 5931269
23-05-2014
"""
class Tree:

    def __init__(self, root=None):
        self.root = root

    def set_root(self, root):
        self.root = root

    def get_root(self):
        return self.root

    def get_leaves(self):
        return self.root.get_leaves()

    def get_leaf_count(self):
        return self.root.get_leaf_count()

    def get_newick(self):
        return self.root.get_newick() + ";"

    def __str__(self):
        return self.get_newick()

"""
Python Node Class
Stuart Bradley - 5931269
23-05-2014
"""
class Node:
    def __init__(self, label=None):
        self.parent = None
        self.children = []
        self.height = -1.0
        self.label = label
        self.sequence = None

    def get_parent(self):
        return self.parent

    def set_parent(self, parent):
        self.parent = parent

    def get_children(self):
        return self.children

    def add_child(self, child):
        self.children.append(child)
        child.set_parent(self)

    def remove_child(self, child):
        self.children.remove(child)

    def set_height(self, height):
        self.height = height

    def get_height(self):
        return self.height

    def is_root(self):
        return self.parent == None

    def is_leaf(self):
        return not self.children

    def get_sequence(self):
        return self.sequence

    def set_sequence(self, sequence):
        self.sequence = sequence

    def get_label(self):
        return self.label

    def set_label(self, label):
        self.label = label

    def get_leaves(self):
        leaf_list = []

        if (self.is_leaf()):
            leaf_list.append(self)
        else:
            for child in self.children:
                leaf_list.extend(child.get_leaves())

        return leaf_list

    def get_leaf_count(self):
        if self.is_leaf():
            return 1
        else:
            return sum(map(Node.get_leaf_count, self.children))

    def get_newick(self):
        sb = ""

        if (not self.is_leaf()):
            sb += "("
            for i in range(0, len(self.children)):
                if (i>0):
                    sb += ","
                sb += self.children[i].get_newick()
            sb += ")"

        if (self.label != None):
            sb += self.label

        branch_length = -1.0
        if (not self.is_root()):
            branch_length = self.parent.height - self.height
        else:
            branch_length = 0.0

        sb += ":" + str(branch_length)

        return sb

    
# Calculate the UPGMA tree from the distance matrix
tree = compute_upgma_tree([[0, 4, 8, 8], [4, 0, 8, 8], [8, 8, 0, 6], [8, 8, 6, 0]])

# Plot the tree
plot_tree(tree)
```

# Question 1
Kingman Coalescent Method

```python
import math
import random
import numpy as np

# Creates a random tree based on n the number of leaves
# and Ne the effective population size. Returns the root of a Tree
def kingman(n, Ne):
    # initalize
    k, t = n, 0
    current_nodes = []
    theory = theoretical_mean(n, Ne)
    
    # create n nodes labelled from 1 to n
    for i in range(1, n + 1):
        new_node = Node(str(i))
        new_node.set_height(t)
        current_nodes += [new_node]
        
    # Loop which matches up random pairs of leaves
    while (k > 1):
        # Calculate a random time based on an exponential distribution
        tk = np.random.exponential(1/(C(k, 2) / Ne))
        
        # Add the random time to current time
        t += tk
        # Select two nodes at random from available nodes. Remove them from the available nodes list
        # Make i, j children of a node
        current_node = create_parent(choose_remove(current_nodes), choose_remove(current_nodes))
        current_node.set_height(t)
        # add the node back to the list of available nodes
        current_nodes += [current_node]

        k -= 1
    
    return Tree(current_nodes[0])

#########################################################
# Supplmentary functions to assist the kingman function #
#########################################################    

# Takes two nodes and returns a node which is parent to the two nodes
def create_parent(node_a, node_b):
    current_label = "({0} - {1})".format(node_a.get_label(), node_b.get_label())
    current_node = Node(current_label)
    current_node.add_child(node_a)
    current_node.add_child(node_b)
    return current_node

# chooses a node from a node list at random,
# removes it from the list and returns it
def choose_remove(node_list):
    random_node = random.choice(node_list)
    node_list.remove(random_node)
    return random_node

# n choose r
def C(n, r):
    top = math.factorial(n)
    bottom = (math.factorial(r)*math.factorial(n-r))
    return top/bottom

# calculates the Theoretical mean of two values and returns it as a float
def theoretical_mean(n, Ne):
    return 2 * Ne * (1-(1/n))

# takes a tree, and prints it. Returns nothing
def print_tree(a_tree):
    root_node = a_tree.get_root()
    # print this Tree
    print("\nPlotting Tree")
    plot_tree(a_tree)

# produces x trees, and prints the average height of the trees, and returns the last tree simulated
def simulate_trees(trees_to_produce, n, Ne):
    height_sum = 0
    x = 0
    # Create random trees and calculate the height of the root node
    while (x < trees_to_produce):
        a_tree = kingman(n, Ne)
        curr_height = a_tree.get_root().get_height()
        height_sum += curr_height
        x += 1

    # Print out summary information 
    if (x == trees_to_produce):
        print("Mean height of {1} samples: {0} \nHeight of the last tree: {2}".format(height_sum / trees_to_produce, x, curr_height))
        plot_tree(a_tree)
    return a_tree


def question_1():
    #init
    n, Ne = 10, 100
    # Run 1000 trees and prints summary information
    simulate_trees(1000, n, Ne)

question_1()
```

I have created a function which loops 1000 times to calculate that my mean height was close to the theoretical mean to ensure correctness of the algorithm.

Finally I have print the last tree created with height on the x axis, and leaves branches on the y axis.

Jukes Cantor Substitution model

```python
import copy;

# When given a tree, creates a random sequence, which is stored in the root
# Traverses down the children of each node and mutates the
# sequence based on mu L and T.
# Returns the tree with sequence information stored on nodes
def jukes_cantor(tree, sequence_length, mutation_rate):
    # Set the root
    current_node = tree.get_root()
    current_node.set_sequence(rand_seq(sequence_length))
        
    # If children exist, visit them
    children = current_node.get_children()
    if (children):
        for child in children:
            traverse(child, sequence_length, mutation_rate)
    
    return tree;

# Traverses a tree and recursively prints out a sequence of genome sequences
# from the root to each leaf. Nothing returned
def print_mutation_tree(a_tree):
    list_to_print = []
    root_sequence = sequence_to_str(a_tree.get_root().get_sequence())
    root_sequence = "{} || root sequence".format(root_sequence)
    list_to_print += [root_sequence]
    
    children = a_tree.get_root().get_children()
    if (children):
        for child in children:
            list_to_print = branch_printer(child, root_sequence, list_to_print)
            
# Critical function of print_mutation_tree
def branch_printer(a_node, root_sequence, list_to_print):
    list_to_print += ["{1} || Node label: {0}: ".format(a_node.get_label(), sequence_to_str(a_node.get_sequence()))]
    if a_node.is_leaf():
        print("-" * len(root_sequence))
        if len(list_to_print):
            for string in list_to_print:
                print(string)
    else:
        children = a_node.get_children()
        if children:
            for child in children:
                list_to_print = branch_printer(child, root_sequence, list_to_print)
    
    list_to_print = list_to_print[:-1]
    return list_to_print

# Visits every child of a given node, mutates the given node
def traverse(node, sequence_length, mutation_rate):    
    # Mutate the parent sequence and assign to this node
    parent_seq = node.get_parent().get_sequence()
    time = node.get_parent().get_height() - node.get_height()
    node.set_sequence(copy.deepcopy(mutate(parent_seq, time, mutation_rate)))
    
    # Recursive call on the children
    if not(node.is_leaf()):
        children = node.get_children()
        if (children):
            for child in children:
                traverse(child, sequence_length, mutation_rate)

    
# produces a random genome or length L
# and returns that random genome
def rand_seq(L):
    seq = []
    for i in range(L):
        seq += random.choice("ACGT")
    return seq

# Given a sequence, time and mutation rate,
# mutates that sequence based on the mutation rate
# and time. Returns the mutated sequence
def mutate(X,t,mu):
    L = len(X)
    # the number of mutations is Poisson with total rate L*mu*t
    numMutation = np.random.poisson(L*mu*t)
    # for each mutation, choose a site to mutate and mutate it
    for i in range(numMutation):
        # choose a site
        site = math.floor(random.random()*L)
        # mutate that site
        X[site] = random.choice("ACGT")
    return X

# Takes a sequence list and outputs it as a string
def sequence_to_str(sequence):
    out_str = ""
    for char in sequence:
        out_str += char
    return out_str
```

# Question 3
Jukes Cantor Distance Matrix

```python
# Distance function. Takes a list of sequences and returns a matrix
# that stores the distance between each sequence
def create_distance_matrix(sequence_list):
    numpy_matrix = []
    for i in range(len(sequence_list)):
        numpy_row = []
        for j in range(len(sequence_list)):
            curr_value = dxy(sequence_list[i], sequence_list[j])
            numpy_row += [curr_value]
        numpy_matrix += [numpy_row]
    return numpy_matrix

#####################################################################
# Supplmentary functions to assist the Jukes-Cantor distance matrix #
#####################################################################

# Jukes-Cantor Distance Matrix Function
# Takes a tree and returns a distance matrix and a sequence list
def jc_distance_matrix(in_tree):
    label_dict = {}
    grab_sequences(in_tree.get_root(), label_dict)
    sequence_list = [label_dict[i] for i in range(1,len(label_dict) + 1)]
    distance_matrix = create_distance_matrix(sequence_list)
    return distance_matrix, sequence_list

# Takes a root_node and a sequence matrix and adds the 
# leaves of the tree to the sequence matrix
def grab_sequences(in_node, label_dict):
    if (in_node.is_leaf()):
        current_sequence = in_node.get_sequence()
        current_label = int(in_node.get_label())
        label_dict[current_label] = current_sequence
    else:
        children = in_node.get_children()
        for child in children:
            grab_sequences(child, label_dict)

# Jukes-Cantor distance Function.
# Takes two strings and returns a log(difference)
def dxy(string_x, string_y):
    return (-3/4)*math.log(1-(4 * fxy(string_x, string_y))/3)

# Fraction of differing sites between two strings
# Returns a float
def fxy(string_x, string_y):
    L = len(string_x)
    return min(Dxy(string_x, string_y, L)/L, 0.75-(1/L))
    
# Counts the number of differences between two strings
# Returns an int
def Dxy(string_x, string_y, L):
    differences = 0
    for i in range(L):
        if (string_x[i] != string_y[i]):
            differences += 1
    return differences

# Takes a distance matrix as an input and prints it
def print_dist(distance_matrix):
    for i in range(len(distance_matrix)):
        for j in range(len(distance_matrix[i])):
            out_str = format_cell(str(round(distance_matrix[i][j], 3)))
            if (j != len(distance_matrix[i]) - 1):
                print(out_str, end=", ")
            else:
                print(out_str)
    print()
        
# Modifies distances so they are strings of consistent length
def format_cell(a_str):
    if len(a_str) == 1:
        a_str = " " + a_str
    while (len(a_str)< 6):
        a_str = a_str + "0"
    return a_str
        
```

```python
def question_4():
    # Init
    n = 10
    Ne = 100
    mu = 0.0015
    length_list = [50, 200, 1000]
    
    # Generate a tree
    a_tree = kingman(n, Ne)
    root_node = a_tree.get_root()
    
    for L in length_list:
        # Populate the tree with sequences of length L
        jukes_cantor(a_tree, L, mu)
        # Calculate the distance matrix
        distance_matrix, sequence_list = jc_distance_matrix(a_tree)
        
        # Title Generator
        title = "Sequence Length: {0}  n: {1}  Ne: {2}  mu: {3}".format(L, n, Ne, mu)
        border = len(title)
        print("-"*border)
        print(title, end= "\n")
        print("-"*border, end = "\n\n")

        # Print Sequence information
        for i in range(n):
            print("Label: {0}  Sequence: \n{1}".format(i + 1, sequence_to_str(sequence_list[i])))

        # Print the distance matrix to 4 dp
        print("\nDistance Matrix\n")
        print_dist(distance_matrix)
    
        # Prints the original tree
        print("\nOriginal Tree")
        plot_tree(a_tree)
        
        # Prints the trees
        print("Reconstructed Tree")
        plot_tree(compute_upgma_tree(distance_matrix))
    # Returns a tree for question 5
    return a_tree


a_tree = question_4()
```

```python
import matplotlib.pyplot as plt
import statistics
from scipy.stats import lognorm

def p_T_Ne(n, Ne, tk_list):
    total = 0
    i = 0
    for k in range(2, n + 1):
        total += -(math.log(Ne)) - ( (k * (k - 1) * tk_list[i]) / (2 * Ne) )
        i += 1
    return total

def accept(pre_rate, post_rate):
    try: 
        accept_rate = 1 - min(1, math.exp(pre_rate - post_rate))
    except:
        accept_rate = 0
    rand = np.random.uniform(0,1)
    if (accept_rate > rand):
        return True
    else:
        return False
    
# takes a node and a list mutates the list to append 
# the height of the current node and then visits its children
def get_all_heights(a_node, height_list):
    height = a_node.get_height()
    height_list += [height]
    children = a_node.get_children()
    for child in children:
        get_all_heights(child, height_list)
    

# Takes a list of heights and creates list of 
# differences between a_list[i] - a_list[i + 1]
# for i_1 = 0 to len(a_list)
def create_tk_list(a_list):
    a_list.sort()
    a_list.reverse()
    out_list = []
    for i in range(len(a_list) - 1):
        out_list += [a_list[i] - a_list[i+1]]
    return out_list
```

```python
def question_5(a_tree, step_size=10):
    # int assignments
    n = 10
    Ne = 1
    steps = 10000
    running_sum = 0
    count = 0
    # list assignments
    height_list = [] #list of all the heights
    Ne_list = [Ne]
    mean_list = []
    std_dv_list = []
    
    # populate lists
    root_node = a_tree.get_root()
    get_all_heights(root_node, height_list)
    height_list = create_tk_list(height_list)
    
    # Algorithm start
    for i in range(steps):
        # Generate a proposed Ne'
        Ne = Ne_list[-1]
        random_step = math.floor(np.random.uniform(-step_size, step_size + 1))
        Ne += random_step
        # reflect Ne if negative
        if (Ne < 0):
            Ne = abs(Ne)
        elif (Ne == 0):
            Ne = 1
        
        # Generate two probabilities for Ne and Ne'
        # Prior is a lognorm sample using the running std_dev and mean
        # For computational efficiency, std dev is only calculated ever
        chance_old_Ne = p_T_Ne(n, Ne_list[-1], height_list) + np.random.lognormal()
        chance_new_Ne = p_T_Ne(n, Ne, height_list) + np.random.lognormal()
        #calculate the chance to accept Ne'
        acceptance = accept(chance_old_Ne, chance_new_Ne)
        if (not(acceptance)):
            Ne_list += [Ne_list[-1]]
            running_sum += Ne_list[-1]
        else:
            count += 1
            Ne_list += [Ne]
            running_sum += Ne
        current_mean = running_sum / (i + 1)
        mean_list += [current_mean]
        
        # Sample the standard deviation of Ne_list every 100 items
        # the more often you run this the slower the program will be
        if i%100 == 0 or i < 10:
            std_dv_list += [statistics.stdev(Ne_list)]
        else:
            std_dv_list += [std_dv_list[-1]]
    
    # Creates a list of proposed values for Ne by sampling a normal distribution
    print_dist = [math.floor(np.random.lognormal() * mean_list[-1] + np.random.uniform(-std_dv_list[-1], std_dv_list[-1]+1)) for i in range(steps)]
    std_dv = statistics.stdev(Ne_list)
    
    print("step_size = {}".format(step_size))
    print("Real Ne: {}".format(a_tree.get_root().get_height()))
    
    plt.plot(Ne_list)
    plt.plot(std_dv_list)
    plt.plot(mean_list)
    plt.ylabel('Ne')
    plt.xlabel("Steps")
    plt.title("Ne vs Steps")
    plt.show()
    
    # Print histogram
    print_dict = {}
    for i in range(300):
        print_dict[i] = 0
        
    for Ne in Ne_list:
        try:
            print_dict[Ne] += 1
        except:
            pass
    print_hist = []
    
    for i in range(250):
        print_hist += [print_dict[i]]
        
    
    dist_dict = {}
    for i in range(300):
        dist_dict[i] = 0
    for rand in print_dist:
        try:
            print_dict[math.floor(rand)] += 1
        except:
            pass
    print_dist = []
    for i in range(250):
        print_dist += [print_dict[i]]
    
    plt.plot(print_dist)
    plt.plot(print_hist)
    plt.title("Distribution of Ne")
    plt.xlabel('Ne')
    plt.ylabel('Count')
    plt.show()
    
    print("hit %: {0}".format(100 * count / steps))
    print("mean Ne: {}\n".format(mean_list[-1]))
    
    return mean_list[-1]

    

question_5(a_tree)
```
